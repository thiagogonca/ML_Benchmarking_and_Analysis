```python
import pandas as pd

data_path = "../../../data"

prc_vazao = f"{data_path}/vazao/processed"
prc_merge = f"{data_path}/merge"
datasets = f"{data_path}/datasets"
```

```python
vazao_diaria = pd.read_csv(f'{prc_vazao}/vazao_diaria.csv', index_col='DATA') # aqui cada usina tem um range diferente
vazao_diaria
```

```python
usina = 'ITAIPU (266)'
itaipu = vazao_diaria[[usina]]
itaipu
```

```python
bacia_parana = pd.read_csv(f'{prc_merge}/MERGE_PREC_BACIA_ITAIPU_DAILY_2000_2023.csv', 
                           index_col='data').drop(columns=['index'])
bacia_parana 
```

```python
df = bacia_parana.join(itaipu, how='inner').rename(
    columns={'prec':'bacia_prec_sum', f'{usina}':'vazao_itaipu'})

df.index = pd.to_datetime(df.index)
df
```

```python
df = df.resample('M').sum()
df
```

```python
df.to_csv(f'{datasets}/Itaipu_POC_VAZAO_V2.csv', index_label='time')
pd.read_csv(f'{datasets}/Itaipu_POC_VAZAO_V2.csv', index_col='time') 
```
